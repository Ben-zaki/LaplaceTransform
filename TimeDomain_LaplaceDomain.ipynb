```python
from sympy import *
```

```python
t, s = symbols('t s',positive = True)
F = 5*t
U = laplace_transform(F,t,s)#from time domain to s domain
U
```

```python
U[0]
```

```python
inverse_laplace_transform(U[0], s, t).evalf().simplify()#from s domain to time domain
```

### Factorisation

```python
# Using sympy.apart() method to factorize a function
s = symbols('s')
apart((s**2 + 10 * s + 16)/(s**2 + s))  
```

```python
s, t = symbols("s t",positive=True)
H = (2*s+1)/(s**3 + 2*s**2 + 5*s)#laplace transfrom of some function h to be inverted
inverse_laplace_transform(H,s,t).evalf().simplify()#renders inverse laplace transform
```

```python
s, t = symbols("s t", positive=True)
h = 150*s/((s + 2)*(3*s + 1))
inverse_laplace_transform(h,s,t).evalf().simplify()
```
